# BMI Calculator

```python
name = input("Enter your name: ")
while True:
        weight_unit = input("Enter unit of weight(lb or kg): ").lower().strip()
        if weight_unit == "lb":
            while True:
                try:
                    weight_pounds = float(input("Enter your weight in pounds: "))
                    if weight_pounds > 0:
                        while True:
                            height_inches = float(input("Enter your height in inches: "))
                            if height_inches > 0:
                                BMI_pounds = round(float((weight_pounds * 703) / (height_inches ** 2)), 1)
                                print(f"{name}, your BMI is {BMI_pounds}")
                                if BMI_pounds < 18.5:
                                    print(f"{name}, you are underweight")
                                elif 18.5 <= BMI_pounds <= 24.9:
                                    print(f"{name}, you have normal weight")
                                elif 25.0 <= BMI_pounds <= 29.9:
                                    print(f"{name}, you are overweight.")
                                elif 30.0 <= BMI_pounds <= 34.9:
                                    print(f"{name}, you are obese")
                                elif 35.0 <= BMI_pounds <= 39.9:
                                    print(f"{name}, you are severely obese.")
                                elif BMI_pounds >= 40.0:
                                    print(f"{name}, you are morbidly obese.")
                                break
                            else:
                                print("Please enter a positive number!")
                                continue
                        break
                    else:
                        print("Please enter a positive number!")
                        continue
                except ValueError:
                    print("Please enter a number!")
                    pass
            break
        elif weight_unit == "kg":
            while True:
                try:
                    weight_kilos = float(input("Enter your weight in kilograms: "))
                    if weight_kilos > 0:
                        while True:
                            try:
                                height_meters = float(input("Enter your height in meters: "))
                                if height_meters > 0:
                                    BMI_kilos = round(float(weight_kilos / (height_meters ** 2)), 1)
                                    print(f"{name}, your BMI is {BMI_kilos}")
                                    if BMI_kilos < 18.5:
                                        print(f"{name}, you are underweight")
                                    elif 18.5 <= BMI_kilos <= 24.9:
                                        print(f"{name}, you have normal weight")
                                    elif 25.0 <= BMI_kilos <= 29.9:
                                        print(f"{name}, you are overweight.")
                                    elif 30.0 <= BMI_kilos <= 34.9:
                                        print(f"{name}, you are obese")
                                    elif 35.0 <= BMI_kilos <= 39.9:
                                        print(f"{name}, you are severely obese.")
                                    elif BMI_kilos >= 40.0:
                                        print(f"{name}, you are morbidly obese.")
                                    break
                                else:
                                    print("Please enter a positive number!")
                                    continue
                            except ValueError:
                                print("Please enter a number!")
                                pass
                        break
                    else:
                        print("Please enter a positive number!")
                        continue
                except ValueError:
                    print("Please enter a number!")
                    pass
            break
        else:
            print("Please enter a valid unit!")
            continue
```
